In the first lecture we considered the general qubit Hamiltonian
$$\hat{H} = \frac{\hbar\omega}{2} \mathbf{u} \cdot \hat{\boldsymbol{\sigma}}$$
for some unit vector $\mathbf{u}= (u_x,u_y,u_z)$. In this exercise we will find a general analytical expression for the time evolution operator $ \hat{U}(t) = {\rm e}^{-{\rm i}\hat{H}t/\hbar}$

(a) Let's warm up with the simplest case where $\mathbf{u} = (0,0,1)$. First show that $\hat{\sigma}_z^2 = \hat{\mathbb{1}}$ is equal to the identity operator. Now use this fact and the definition of the matrix exponential (i.e. its Taylor expansion) to show that 

$$ {\rm e}^{-{\rm i}\omega \hat{\sigma}_zt/2} = \cos(\omega t/2) \hat{1} - {\rm i}\sin(\omega t/2) \hat{\sigma}_z.$$

Hint: recall the Taylor expansions for $\cos(x) = \sum_{n=0}^\infty \frac{({\rm i} x)^{2n}}{(2n)!}$ and ${\rm i}\sin(x) = \sum_{n=0}^\infty \frac{({\rm i}x)^{2n+1}}{(2n+1)!}$

#### Solution:

The pauli matrix $\hat{\sigma}_z$ is defined as $ \hat{\sigma}_z = \begin{pmatrix} 1 & 0 \\\ 0 & -1 \end{pmatrix}$

We can simply express $\hat{\sigma}_z^2 = \hat{\sigma}_z * \hat{\sigma}_z = \begin{pmatrix} 1 & 0 \\\ 0 & -1 \end{pmatrix} * \begin{pmatrix} 1 & 0 \\\ 0 & -1 \end{pmatrix} $

And we finally get the product: $\begin{pmatrix} 1 & 0 \\\ 0 & 1 \end{pmatrix}$ which is the identity operator $\hat{\mathbb{1}}$

For the second section of this part, we can expand the exponential function ${\rm e}^{-{\rm i}\omega \hat{\sigma}_zt/2}$ to its taylor polynomial:

$T(x) = 1 + {-{\rm i}\omega \hat{\sigma}_zt/2} + \frac{({-{\rm i}\omega \hat{\sigma}_zt/2})^2}{2!} + \frac{({-{\rm i}\omega \hat{\sigma}_zt/2})^3}{3!}+ \frac{({-{\rm i}\omega \hat{\sigma}_zt/2})^4}{4!} + . . . $

We can simplify this down

$T(x) = 1 + {-{\rm i}\omega \hat{\sigma}_zt/2} + \frac{{-\omega^2 (t/2)^2}}{2!} + \frac{{{\rm i}\omega^3 \hat{\sigma}_z(t/2)^3}}{3!}+ \frac{({\omega^4 (t/2)^4})}{4!} + . . . $

We start seeing the similarities to the taylor expansions of $\sin{}$ and $\cos{}$.

$T(x) = [1 + \frac{{-\omega^2 (t/2)^2}}{2!}  + \frac{({\omega^4 (t/2)^4})}{4!} + . . .] + [{-{\rm i}\omega \hat{\sigma}_zt/2}+ \frac{{{\rm i}\omega^3 \hat{\sigma}_z(t/2)^3}}{3!}+ . . .]  $

$T(x) = [1 + \frac{{-\omega^2 (t/2)^2}}{2!}  + \frac{({\omega^4 (t/2)^4})}{4!} + . . .] + -{\rm i}\hat{\sigma}_z[{(\omega t/2)}- \frac{{(t/2)^3}}{3!}+ . . .]  $

$T(x) = \cos{(\omega t/2)}- {\rm i}\hat{\sigma}_z \sin{(\omega t/2)}$

${\rm e}^{-{\rm i}\omega \hat{\sigma}_zt/2} \approx \cos{(\omega t/2)}- {\rm i}\hat{\sigma}_z \sin{(\omega t/2)}$

(b) The commutator between two operators $\hat{A}$ and $\hat{B}$ is defined by $[\hat{A},\hat{B}] = \hat{A}\hat{B}-\hat{B}\hat{A}$. The anti-commutator is defined by $\{\hat{A},\hat{B}\} = \hat{A}\hat{B}+\hat{B}\hat{A}$. The commutator between any pair of Pauli matrices is given in the typed lecture notes ([see Blackboard](https://tcd.blackboard.com)). Convince yourself that the anti-commutation relation

$$\{\hat{\sigma}_j,\hat{\sigma}_k \} = 2\delta_{jk} \hat{\mathbb{1}}$$

holds for any pair of indices $j,k = x,y,z$. Here, $\delta_{jk}$ is the Kronecker delta symbol, defined by $\delta_{jk} = 1$ if $j=k$ and $\delta_{jk} = 0$ otherwise. Using these 
commutation and anti-commutation relations (or otherwise), verify the identity:

$$ \hat{\sigma}_j\hat{\sigma}_k = \delta_{jk}  \hat{\mathbb{1}} + \sum_{l=x,y,z}{\rm i} \varepsilon_{jkl} \hat{\sigma}_l,$$

where $\varepsilon_{ijk}$ is the Levi-Civita symbol, which is anti-symmetric under any pairwise permutation of its indices, $\varepsilon_{jkl}= -\varepsilon_{kjl} = \varepsilon_{klj}\ldots $ etc., and takes the value $\varepsilon_{jkl} = 1$ for $(j,k,l) = (x,y,z)$. Note that the anti-symmetric property implies that $\varepsilon_{jkl}=0$ whenever two or more indices are equal.

#### Solution:
- Resource for better understanding pauli matrices: [link here](https://en.wikipedia.org/wiki/Pauli_matrices)

Anticommutation relation: $ \hat{\sigma}_i \hat{\sigma}_j + \hat{\sigma}_j \hat{\sigma}_i = 2 \delta_{ij} \hat{\mathbb{1}} $

Commutation relation: $ \hat{\sigma}_i \hat{\sigma}_j - \hat{\sigma}_j \hat{\sigma}_i = 2\rm i \varepsilon_{ijk}\hat{\sigma}_l$

Identity: $\hat{\sigma}_j \hat{\sigma}_j = \hat{\mathbb{1}} \quad$

The product $\hat{\sigma}_i \hat{\sigma}_j$ can be expressed as $\frac{1}{2}(\hat{\sigma}_i \hat{\sigma}_j + \hat{\sigma}_j \hat{\sigma}_i) - \frac{1}{2}(\hat{\sigma}_i \hat{\sigma}_j - \hat{\sigma}_j \hat{\sigma}_i)$

We then substitue the results of the anticommutation and commutation relations to get: 

$\hat{\sigma}_j\hat{\sigma}_k = \delta_{jk}  \hat{\mathbb{1}} + \sum_{l=x,y,z}{\rm i} \varepsilon_{jkl} \hat{\sigma}_l$


(c) Now consider a general $\mathbf{u}=(u_x,u_y,u_z)$, where all three components of $\mathbf{u}$ are non-zero. Using the identity from (b) and the fact that $\lVert\mathbf{u}\rVert = 1$, prove that

$$ \hat{H}^2 = \left(\frac{\hbar\omega}{2}\right)^2\hat{\mathbb{1}}.$$

Use this result to show that, in general,

$$ \hat{U}(t) = \cos(\omega t/2) \hat{\mathbb{1}} -{\rm i} \sin(\omega t/2) \mathbf{u} \cdot \hat{\boldsymbol{\sigma}}.$$

#### Solution:
- Resource for better understanding pauli matrices: [link here](https://en.wikipedia.org/wiki/Pauli_matrices)

We consider the general qubit hamiltonian
$ \hat{H} = \frac{\hbar\omega}{2} \mathbf{u} \cdot \hat{\boldsymbol{\sigma}} $

We then can square $\hat{H}$ so that $\hat{H}^2 = (\frac{\hbar\omega}{2} \mathbf{u} \cdot \hat{\boldsymbol{\sigma}})^2 = ((\frac{\hbar\omega}{2})^2 (\mathbf{u} \cdot \hat{\boldsymbol{\sigma}})\cdot(\mathbf{u} \cdot \hat{\boldsymbol{\sigma}}))$

We know that in general, $ \hat{A}_j \hat{A}_k = \delta_{jk} \hat{\mathbb{1}} + \sum_{l=x,y,z} \varepsilon_{ijk} \hat{A}_l$, and that for different vectors, the result of the delta will be 0, which we can use to simplify our equation. We also can use the fact that $\lVert\mathbf{A}\rVert$ will return $\hat{\mathbb{1}}$

We also know that in general $ \hat{A} \hat{B} = (\hat{A} \cdot \hat{B})\hat{\mathbb{1}} + \rm i (\hat{A} \times \hat{B})$. Let's use this for the equation:

$\hat{H}^2 = (\frac{\hbar\omega}{2} \mathbf{u} \cdot \hat{\boldsymbol{\sigma}})^2 = ((\frac{\hbar\omega}{2})^2 ((\mathbf{u} \cdot \mathbf{u})\hat{\mathbb{1}} + \rm i \hat{\boldsymbol{\sigma}} (\mathbf{u} \times \mathbf{u} ))$

We understand that cross product the same matrix returns 0 and from the previous part that $\mathbf{u} \cdot \mathbf{u}$, or $\lVert\mathbf{u}\rVert$ will return $\hat{\mathbb{1}}$

So we finally get that $\hat{H}^2 = (\frac{\hbar\omega}{2})^2 \hat{\mathbb{1}} $


Earlier it was defined to us that the time evolution operator is $ \hat{U}(t) = {\rm e}^{-{\rm i}\hat{H}t/\hbar}$. 

We can substitute the general qubit hamiltonian $ \hat{H} = \frac{\hbar\omega}{2} \mathbf{u} \cdot \hat{\boldsymbol{\sigma}} $ into $\hat{H}$ in the time evolution operator.

$ \hat{U}(t) = {\rm e}^{-{\rm i}\frac{\omega}{2} \mathbf{u} \cdot \hat{\boldsymbol{\sigma}}t}$, we then can use the taylor expansions. 

$ \hat{U}(t) \approx 1 + -{\rm i}\frac{\omega}{2} \mathbf{u} \cdot \hat{\boldsymbol{\sigma}}t + \frac{(-{\rm i}\frac{\omega}{2} \mathbf{u} \cdot \hat{\boldsymbol{\sigma}}t)^2}{2!} + \frac{(-{\rm i}\frac{\omega}{2} \mathbf{u} \cdot \hat{\boldsymbol{\sigma}}t)^3}{3!} + \frac{(-{\rm i}\frac{\omega}{2} \mathbf{u} \cdot \hat{\boldsymbol{\sigma}}t)^4}{4!} + ...$

We know in the previous part that (($\mathbf{u} \cdot \boldsymbol{\sigma}))^2 = \hat{\mathbb{1}}$, so for the even parts of the taylor expansion we can replace the dot product with 1.

$ \hat{U}(t) \approx 1 + {\rm i}\frac{\omega}{2}t(\mathbf{u}\cdot \hat{\boldsymbol{\sigma}}) + \frac{-(\frac{\omega}{2}t)^2}{2!} + \frac{{\rm i}(\frac{\omega}{2}t)^3(\mathbf{u} \cdot \boldsymbol{\sigma})}{3!} + \frac{(\frac{\omega}{2}t)^4}{4!} + ... $

$ \hat{U}(t) \approx (1 + \frac{-(\frac{\omega}{2}t)^2}{2!} + \frac{(\frac{\omega}{2}t)^4}{4!} + ...) + (-{\rm i}\frac{\omega}{2}t(\mathbf{u}\cdot \hat{\boldsymbol{\sigma}}) + \frac{{\rm i}(\frac{\omega}{2}t)^3(\mathbf{u} \cdot \boldsymbol{\sigma})}{3!} + ...) $

$ \hat{U}(t) \approx \cos(\frac{\omega}{2}t) + -{\rm i}(\mathbf{u}\cdot \hat{\boldsymbol{\sigma}})\sin(\frac{\omega}{2}t) \approx \hat{U}(t) \approx \cos(\frac{\omega}{2}t) - {\rm i}(\mathbf{u}\cdot \hat{\boldsymbol{\sigma}})\sin(\frac{\omega}{2}t)$

(d) Set $\mathbf{u} = (1,1,0)/\sqrt{2}$, and use the result of (c) to calculate an analytical expression for the operator $\hat{\sigma}_z$ in the Heisenberg picture:

$$ \hat{\sigma}_{z,H}(t) = \hat{U}^\dagger(t) \hat{\sigma}_z \hat{U}(t). $$

#### Solution:

We know that $ \hat{U}(t) =\cos(\frac{\omega}{2}t) - {\rm i}(\mathbf{u}\cdot \hat{\boldsymbol{\sigma}})\sin(\frac{\omega}{2}t)$, so we can set the transpose conjugate $\hat{U}^\dagger(t)$ as $\hat{U}^\dagger(t) = \cos(\frac{\omega}{2}t) + {\rm i}(\mathbf{u}\cdot \hat{\boldsymbol{\sigma}})^*\sin(\frac{\omega}{2}t)$

We then can substitute $ \hat{U}(t)$ and $\hat{U}^\dagger(t)$ into $ \hat{\sigma}_{z,H}(t) = \hat{U}^\dagger(t) \hat{\sigma}_z \hat{U}(t). $

$[\cos(\frac{\omega}{2}t) + {\rm i}(\mathbf{u}\cdot \hat{\boldsymbol{\sigma}})^*\sin(\frac{\omega}{2}t)]\hat{\sigma}_z[\cos(\frac{\omega}{2}t) - {\rm i}(\mathbf{u}\cdot \hat{\boldsymbol{\sigma}})\sin(\frac{\omega}{2}t)]$

Upon solving $\mathbf{u}\cdot \hat{\boldsymbol{\sigma}}$ using Eistein's summation convention, we get $\mathbf{u}\cdot \hat{\boldsymbol{\sigma}} = \frac{1}{\sqrt{2}}\begin{pmatrix} 0 & 1-{\rm i} \\\ 1+{\rm i} & 0 \end{pmatrix}$. Upon further inspection this matrix is the sum of $\hat{\sigma}_x$ and $\hat{\sigma}_y$ multiplied by $\frac{1}{\sqrt{2}}\$.

So let's substitute $\mathbf{u}\cdot \hat{\boldsymbol{\sigma}}$ with $\hat{\sigma}_x + \hat{\sigma}_y$, then distribute $\hat{\sigma}_z$ to the left side

$[\cos(\frac{\omega}{2}t) + {\rm i}(\frac{1}{\sqrt{2}}(\hat{\sigma}_x + \hat{\sigma}_y))^*\sin(\frac{\omega}{2}t)]{\sigma}_z[\cos(\frac{\omega}{2}t) - {\rm i}(\frac{1}{\sqrt{2}}(\hat{\sigma}_x + \hat{\sigma}_y))^*\sin(\frac{\omega}{2}t)]$

$[\cos^2(\frac{\omega}{2}t)\hat{\sigma}_z + \frac{{\rm i}}{\sqrt{2}}(\hat{\sigma}_x\hat{\sigma}_z + \hat{\sigma}_y\hat{\sigma}_z)\sin(\frac{\omega}{2}t)][\cos^2(\frac{\omega}{2}t)\hat{\sigma}_z - \frac{{\rm i}}{\sqrt{2}}(\hat{\sigma}_x\hat{\sigma}_z + \hat{\sigma}_y\hat{\sigma}_z)\sin(\frac{\omega}{2}t)]$

We can then multiply and add the matrices.

$[\cos(\frac{\omega}{2}t)\begin{pmatrix} 1 & 0 \\\ 0 & -1 \end{pmatrix}+\frac{{\rm i}}{\sqrt{2}}\begin{pmatrix} 0 & 1-{\rm i} \\\ 1+{\rm i} & 0 \end{pmatrix}\sin(\frac{\omega}{2}t)][\cos(\frac{\omega}{2}t)-\frac{{\rm i}}{\sqrt{2}}\begin{pmatrix} 0 & 1-{\rm i} \\\ 1+{\rm i}  & 0 \end{pmatrix}\sin(\frac{\omega}{2}t)]$

$\cos^2(\frac{\omega}{2}t)\begin{pmatrix} 1 & 0 \\\ 0 & -1 \end{pmatrix} - \frac{\rm i}{\sqrt{2}}\cos(\frac{\omega}{2}t)\sin(\frac{\omega}{2}t)\begin{pmatrix} 0 & 1-{\rm i} \\\ -1-{\rm i} & -1 \end{pmatrix} + \frac{\rm i}{\sqrt{2}}\cos(\frac{\omega}{2}t)\sin(\frac{\omega}{2}t)\begin{pmatrix} 0 & {\rm i}-1 \\\ 1+{\rm i} & 0 \end{pmatrix} - \sin^2(\frac{\omega}{2}t)(\frac{\rm i}{\sqrt{2}})^2\begin{pmatrix} -2 & 0 \\\ 0 & 2 \end{pmatrix}$

$\begin{pmatrix} \cos^2(\frac{\omega}{2}t) & 0 \\\ 0 & -\cos^2(\frac{\omega}{2}t) \end{pmatrix} + \begin{pmatrix} 0 & 2\cos(\frac{\omega}{2}t)\sin(\frac{\omega}{2}t)\frac{\rm i}{\sqrt{2}}({\rm i} - 1) \\\ 2\cos(\frac{\omega}{2}t)\sin(\frac{\omega}{2}t)\frac{\rm i}{\sqrt{2}}({\rm i} + 1) \end{pmatrix} + \begin{pmatrix} -\sin^2(\frac{\omega}{2}t) & 0 \\\ 0 & \sin^2(\frac{\omega}{2}t) \end{pmatrix}$

$\begin{pmatrix} \cos^2(\frac{\omega}{2}t) -\sin^2(\frac{\omega}{2}t) & 2\cos(\frac{\omega}{2}t)\sin(\frac{\omega}{2}t)\frac{\rm i}{\sqrt{2}}({\rm i} - 1) \\\ 2\cos(\frac{\omega}{2}t)\sin(\frac{\omega}{2}t)\frac{\rm i}{\sqrt{2}}({\rm i} + 1) &  -\cos^2(\frac{\omega}{2}t) + \sin^2(\frac{\omega}{2}t) \end{pmatrix}$

We expect a state $\langle \downarrow|\hat{\sigma}_{z,H}(t)|\downarrow\rangle$, so let's find that state.

$\langle \downarrow|\hat{\sigma}_{z,H}(t)|\downarrow\rangle = \begin{pmatrix} 0 & 1 \end{pmatrix} \begin{pmatrix} \cos^2(\frac{\omega}{2}t) -\sin^2(\frac{\omega}{2}t) & 2\cos(\frac{\omega}{2}t)\sin(\frac{\omega}{2}t)\frac{\rm i}{\sqrt{2}}({\rm i} - 1) \\\ 2\cos(\frac{\omega}{2}t)\sin(\frac{\omega}{2}t)\frac{\rm i}{\sqrt{2}}({\rm i} + 1) &  -\cos^2(\frac{\omega}{2}t) + \sin^2(\frac{\omega}{2}t) \end{pmatrix} \begin{pmatrix} 0 \\\ 1 \end{pmatrix}$

$\langle \downarrow|\hat{\sigma}_{z,H}(t)|\downarrow\rangle = \begin{pmatrix} 0 & 1 \end{pmatrix} \begin{pmatrix} 2\cos(\frac{\omega}{2}t)\sin(\frac{\omega}{2}t)\frac{\rm i}{\sqrt{2}}({\rm i} - 1) \\\ -\cos^2(\frac{\omega}{2}t) + \sin^2(\frac{\omega}{2}t) \end{pmatrix} = -\cos^2(\frac{\omega}{2}t) + \sin^2(\frac{\omega}{2}t)$

Using trigonometry, where $\cos^2(x) - \sin^2(x) = \cos(2x)$ we can get an analytical expression:

$-(\sin^2(\frac{\omega}{2}t)-\cos^2(\frac{\omega}{2}t)) = -\cos(2\frac{\omega}{2}t) = -\cos(\omega t)$

```python
import numpy as np
import math
from matplotlib import pyplot as plt

#walkthrough the notebook, i have the work added in through a picture

#initialize with pauli matrices
o_x = np.matrix([[0,1],[1,0]])
o_y = np.matrix([[0,-1j],[1j,0]])
o_z = np.matrix([[1,0],[0,-1]])
#for the thing that looks like number 1
I = np.matrix([[1,0],[0,1]])
#initalize the vector sigma
o = [o_x,o_y,o_z]

wt = np.linspace(0, 4*np.pi, 120) #discretization of time variable(?)
down_state = np.matrix([[0],
                        [1]]) #down state, initial state of qubit

#Ut = cos(wt/2)I - isin(wt/2)u*o
def U(wt):
    #equation provided above!
    Ut = np.cos(wt/2) * I -1j/np.sqrt(2) * np.sin(wt/2)*(o_x+o_y)
    return Ut
def heisenberg_picture(operator, equation):
    #complex conjugate of U * o_z * U
    #getH() function returns complex conjugate
    s = equation.getH() * operator * equation
    return s
#calculate what we want <down| oz,H(t) |down>
def expectation_value(vector, operator):
    #calculate complex conjugate of vector * operator * vector
    v = vector.getH() * operator * vector
    v = complex(v)
    return v
#define the expected function
def f(x):
   return -(np.cos(x)) #define the function
#generate list of expectation values to plot calculated values
values = [expectation_value(down_state, heisenberg_picture(o_z, U(i))) for i in wt]

#create figure
plt.figure(figsize = (11,4), dpi=300)
plt.plot(wt, f(wt), color='red', label= "Expected graph: $ -\cos{\omega t}$")
plt.plot(wt, [i.real for i in values], label='Real num.', c='coral', linewidth=5, alpha=0.5, zorder=2)
#if there's any imaginary value different than 0, plot the imaginary part at all times:
if not [i.imag for i in values]:
    plt.plot(wt, [i.imag for i in values], label='Imag. num.', c='paleturquoise', linewidth = 5, zorder = 1)


#WHAT DOES THE GRAPH SHOW???
#the graph shows the expectation value vs the period (i think)
#expectation value is the average of all possible outcomes weighted by probabilities
#expectation value of the heisenberg picture being in the down state(?)
plt.xlabel(r'$\omega t$')
plt.ylabel(r'$\langle \hat{\sigma}_{z,H} \rangle$')
plt.xticks([np.pi * i for i in range(5)], ['0', r'$\pi$']+[r'$'+str(i+2)+'\pi$' for i in range(3)])
plt.yticks([-1, -0.5, 0, 0.5, 1], [str(i/2 - 1) for i in range(5)])
plt.legend()
plt.tick_params(axis="both", direction="in")
#plt.savefig('1e.pdf', dpi=400)
plt.show()
```
